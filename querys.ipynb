```python
from sqlalchemy import text
```

 Top 5 Most Expensive Products

```python
query=text("""
SELECT Name, `Unit Price USD`, Category, Subcategory 
FROM data 
ORDER BY `Unit Price USD` DESC 
LIMIT 5;
""")      

df=pd.read_sql_query(query,engine)
df
```

Count of Customers Per Country

```python
query=text("""
SELECT Country, COUNT(*) AS total_customers 
FROM data 
GROUP BY Country 
ORDER BY total_customers DESC;
""")
df=pd.read_sql_query(query,engine)
df
```

Find Customers Based on City and State

```python
query=text("""
SELECT Name, City, State 
FROM data 
WHERE City = 'MOUNT BUDD' AND State = 'Western Australia';
           """)
df=pd.read_sql_query(query,engine)
df

```

Find the Most Popular Product Category

```python
query=text("""
SELECT Category, COUNT(*) AS total_entries 
FROM data 
GROUP BY Category 
ORDER BY total_entries DESC 
LIMIT 1;
           """)
df=pd.read_sql_query(query,engine)
df
```

Count of Products by Subcategory

```python
query=text("""
SELECT Subcategory, COUNT(*) AS total_products 
FROM data 
GROUP BY Subcategory 
ORDER BY total_products DESC;
           """)
df=pd.read_sql_query(query,engine)
df
```

List of Customers Who Purchased Products in Specific Categories

```python
query=text("""
SELECT Name, Category, Subcategory 
FROM data 
WHERE Category = 'Cameras and camcorders';
           """)
df=pd.read_sql_query(query,engine)
df
```

List Products Sorted by Color and Price

```python
query=text("""
SELECT Brand, Name, Color, `Unit Price USD` 
FROM data 
ORDER BY Color ASC, `Unit Price USD` DESC;
       """)
df=pd.read_sql_query(query,engine)
df
```

Find the Maximum and Minimum Unit Cost

```python
query=text("""
SELECT MAX(`Unit Cost USD`) AS max_cost, MIN(`Unit Cost USD`) AS min_cost 
FROM data;
""")

df=pd.read_sql_query(query,engine)
df
```

Products from a Specific Brand

```python
query=text("""
SELECT Name, Brand, `Unit Price USD`, Category 
FROM data 
WHERE Brand = 'Contoso';
           """)
df=pd.read_sql_query(query,engine)
df
```
